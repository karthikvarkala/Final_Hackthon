```python
print("Number of rows:", train_data.shape[0])
print("Number of columns:", train_data.shape[1])
```

####Distribution of yield in our data

```python
plt.figure(figsize=(16,6))
# Group the yield data by date and compute the mean yield for each date
yield_by_date = train_data.groupby('date')['yield'].mean()
sns.lineplot(x=yield_by_date.index, y=yield_by_date)
plt.xlabel('Date')
plt.ylabel('Yield')
plt.title('Yield Time Series Plot')
```

```python
# Count the percenatge of each ingredient type in the data
ingredient_type_counts = train_data['ingredient_type'].value_counts()
plt.figure(figsize=(10,8))
plt.pie(ingredient_type_counts, labels=ingredient_type_counts.index, 
        startangle=90, autopct='%1.1f%%',)
plt.axis('equal')
plt.title('Distribution of Ingredient Types')
plt.show()
```

##Insight


* From this we get to know that we have 4 ingredient_types, out of 4 ingredients **majority** of our farms are producing the **ing_w.** So this will be the 1st importent ingredient for our food chain, **2nd** highest is** ing_x**, 3rd** highest is **ing_y**,and **final one ing_z.**   

* our main focus is on ing_w,beacuse it is more used in our food chain

```python
train_data['month'] = pd.to_datetime(train_data['date']).dt.month

# Group by year and ingredient type to get total yield for each year and ingredient type
yield_by_month_and_ingredient = train_data.groupby(['month', 'ingredient_type'])['yield'].sum().reset_index()

# Create a line plot showing yield over time for each ingredient type
fig = px.line(yield_by_month_and_ingredient, x='month', y='yield', color='ingredient_type', title='Total Yield by month and Ingredient Type')
fig.show()
```

```python
#Countplot to display the observations/max occurence in 'farming_company' column
plt.figure(figsize=(15,8))
sns.countplot(y="farming_company", data= train_data )
plt.title("Maximum Occurence of each Farming Company")
plt.ylabel("Farming Companies")
plt.xlabel("Value Count")
plt.xticks(rotation=90)
plt.show()

#Unique value count of the farming_company column
train_data['farming_company'].value_counts(normalize=True)*100
```

#Insight

*  From  this we get to know that we have majorly 5 farming companies  out of 16, which are contributing to our food chain,others are contributing very less.

*  if reqired ingredints we can cantact to obery Farms first. mostly they can solve our issue regarding ingredints.
*  From obery farms it self we are getting 40% of ingredients. 



```python
#Count Plot to display the Unique locations
plt.figure(figsize=(15,8))
sns.countplot(x="deidentified_location", data= train_data)
plt.title("THE UNIQUE LOCATIONS AT WHICH THE FARMS ARE PRESENT.")
plt.xticks(rotation = 45)
plt.show()

# Value count of unique locations
train_data['deidentified_location'].value_counts(normalize=True)*100
```

#Insight



*   From this we get to know that most of the our Farms are located in locations(2532,8421,5489,5410,5290)




```python
# Bar plot to show the average yield of each farming company:
plt.figure(figsize=(15,8))
df = train_data.groupby(['farming_company'])['yield'].mean().reset_index()
sns.barplot(x='farming_company', y='yield', data=df)
plt.title("Average Yield of Each Farming Company")
plt.xticks(rotation=90)
plt.show()
```

##Insight

*   On by seeing average of yield by company ,Northland Organic Foods Corporation and Obery Farms are producing  high amount of yield






```python
# count plot for ing_w location
plt.figure(figsize=(15, 8))
sns.countplot(x='deidentified_location', data=train_data[train_data['ingredient_type'] == 'ing_w'])
plt.title('Count of farms by location for ing_w')
plt.xticks(rotation=90)
plt.show()

# Value count of unique locations
(train_data[train_data['ingredient_type']=='ing_w']['deidentified_location'].value_counts(normalize=True)*100)
```

#insight


*   From this we get know that  from all locations we are getting ing_w, so our major reqirement is easily avalable for us.
*   Mostly we are getting ing_w from the location 5290



```python
#Plot for distribution of Farm Area.
plt.figure(figsize=(18,8))
sns.set(style="darkgrid")
sns.distplot(train_data['farm_area'], kde = True)
plt.xlabel("Farm Area in Square Meters")
plt.ylabel('frequency')
plt.title('Farm Area Distribution')
plt.show()
```

#Insight
* From the above the areas of farms are mostly in the range of (0-20000)

```python
# The Average Farm area of each company?
df = train_data.groupby(['farming_company'])['farm_area'].mean().reset_index()
#df.reset_index()
df.sort_values('farm_area', ascending = False)
# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='farming_company', y='farm_area', data=df, palette='viridis')
plt.title('Average Farm Area by Company')
plt.xlabel('Farming Company')
plt.ylabel('Average Farm Area')
plt.xticks(rotation=90)
plt.show()
```
